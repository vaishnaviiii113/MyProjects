```python
print("Welcome to my computer quiz!")


print("Okay! Let's play...")
print("This quiz games have 3 levels- EASY, MEDIUM, HARD")
print("Let's begin with EASY!")
score = 0

answer = input("How many states are there in India? ")
if answer.lower() == "29":
    print('Correct!')
    score += 1
else:
    print("Incorrect!")

answer = input("Which is the smallest state in India? ")
if answer.lower() == "goa":
    print('Correct!')
    score += 1
else:
    print("Incorrect!")

answer = input("What is the National Currency of India")
if answer.lower() == "rupee":
    print('Correct!')
    score += 1
else:
    print("Incorrect!")

answer = input("Which is the national sport of India? ")
if answer.lower() == "hockey":
    print('Correct!')
    score += 1
else:
    print("Incorrect!")

print("You got " + str(score) + " questions correct!")
print("You got " + str((score / 4) * 100) + "%.")



print("Let's start with Medium level") 


answer = input("Which is the hottest place in India? ")
if answer.lower() == "rajasthan":
    print('Correct!')
    score += 1
else:
    print("Incorrect!")

answer = input("Who was India’s first prime minister? ")
if answer.lower() == "jawahar lal nehru":
    print('Correct!')
    score += 1
else:
    print("Incorrect!")

answer = input("What is the national heritage animal of India? ")
if answer.lower() == "elephant":
    print('Correct!')
    score += 1
else:
    print("Incorrect!")

answer = input("Which of the seven wonders of the world is located in India? ")
if answer.lower() == "taj mahal":
    print('Correct!')
    score += 1
else:
    print("Incorrect!")

print("You got " + str(score) + " questions correct!")
print("You got " + str((score / 4) * 100) + "%.")




print("Let's start with Hard level") 

answer = input("Which Indian state has the smallest coastline? ")
if answer.lower() == "west bengal":
    print('Correct!')
    score += 1
else:
    print("Incorrect!")

answer = input("What is the national information letter of India? ")
if answer.lower() == "white letter":
    print('Correct!')
    score += 1
else:
    print("Incorrect!")

answer = input("Who was India’s first woman Cabinet Minister? ")
if answer.lower() == "rajkumari amrit kaur":
    print('Correct!')
    score += 1
else:
    print("Incorrect!")

answer = input("What is the total area of India?(in square kilometres) ")
if answer.lower() == "3,287,263 square kilometres":
    print('Correct!')
    score += 1
else:
    print("Incorrect!")

print("You got " + str(score) + " questions correct!")
print("You got " + str((score / 4) * 100) + "%.")



print("Thankyou! for playing this game!")
```
